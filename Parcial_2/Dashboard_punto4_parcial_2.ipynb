```python
import gdown
import pandas as pd

# ID del archivo Excel en Google Drive (tómalo de la URL del archivo)
EXCEL_FILE_ID = "1ekmxf6jqaJJwzp8HeZn6pzI_L1DMUUdGgmLCbxt2Eas"

# Descargar el archivo Excel
gdown.download(f"https://drive.google.com/uc?id={EXCEL_FILE_ID}", "audios.xlsx", quiet=False)

# Leer el archivo Excel
df = pd.read_excel("audios.xlsx")

# Obtener la lista de IDs de los audios
audio_ids = df.iloc[:, 0].tolist()  # Primera columna
print("IDs de audio cargados:", audio_ids)
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Procesar cada audio descargado
for audio_id in audio_ids:
    archivo = f"audios/audio_{audio_id}.wav"
    y, sr = librosa.load(archivo, sr=44100)

    fft = np.abs(np.fft.rfft(y))
    frecuencias = np.fft.rfftfreq(len(y), d=1/sr)

    plt.figure(figsize=(10, 4))
    plt.plot(frecuencias, fft)
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Amplitud")
    plt.title(f"Espectro de {archivo}")
    plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo de audio
audio_file = "audio_descargado.wav"  # Sustituye por el nombre de tu archivo
y, sr = librosa.load(audio_file, sr=None)

# Aplicar la Transformada de Fourier para obtener el espectro de frecuencia
D = np.abs(librosa.stft(y))  # STFT: Transformada de Fourier de corto tiempo
plt.figure(figsize=(10, 6))
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), sr=sr, y_axis='log', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title("Espectrograma de la Transformada de Fourier")
plt.show()
```

```python
import librosa.feature

# Extraer los MFCC del archivo de audio
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Mostrar los MFCC
plt.figure(figsize=(10, 6))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.colorbar()
plt.title("MFCC (Coeficientes Cepstrales en las Frecuencias de Mel)")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import os

# Aquí cargamos todos los audios y extraemos características
def extract_features_from_audio(file_path):
    y, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    return np.mean(mfccs, axis=1)  # Promedio de los MFCC para reducir la dimensión

# Cargar los audios y las etiquetas
audio_files = ["audio1.wav", "audio2.wav"]  # Lista de archivos
labels = [0, 1]  # Etiquetas para las clases (por ejemplo, 0 = "palabra1", 1 = "palabra2")

# Extraer características
features = [extract_features_from_audio(file) for file in audio_files]

# Entrenar un modelo
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)

# Usar KNN como clasificador
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Evaluar el modelo
accuracy = model.score(X_test, y_test)
print(f"Precisión del modelo: {accuracy * 100:.2f}%")
```

```python
# Para un nuevo archivo de audio
new_audio_file = "nuevo_audio.wav"  # Reemplaza con tu archivo de prueba
new_features = extract_features_from_audio(new_audio_file)

# Predecir la clase (palabra clave)
predicted_class = model.predict([new_features])
print(f"La palabra clave detectada es: {predicted_class[0]}")
```
